```python
# 1. Outlier person_age (max 144 - vô lý)
df = df[df['person_age'] <= 100]
```

```python
# 2. Outlier person_emp_length (max 123 - vô lý, ko thể làm 123 năm)
df = df[df['person_emp_length'] <= 60]
```

```python
# 3. Fill missing person_emp_length bằng median
df['person_emp_length'] = df['person_emp_length'].fillna(df['person_emp_length'].median())

# 4. Fill missing loan_int_rate theo median của từng loan_grade (lãi suất phụ thuộc grade)
df['loan_int_rate'] = df.groupby('loan_grade')['loan_int_rate'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
# Kiểm tra lại
print(df.isnull().sum().sum())  # phải = 0
print(df.shape)
```

```python
from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:your_password_here@localhost:5432/credit_risk_db')

df.to_sql('loans', engine, if_exists='replace', index=False)
print("Done!")
```
